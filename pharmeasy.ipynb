```python
import requests

data = []

for page in range(1, 50):
    url = f'https://pharmeasy.in/api/otc/getCategoryProducts?categoryId=641&page={page}'
    response = requests.get(url)
    data_json = response.json()
    print(f'{page} page is done')

    all_healthfood = data_json['data']['products']
    for item, healthfood in enumerate(all_healthfood):
        d = {
            'productId': healthfood['productId'],
            'name': healthfood['name'],
            'manufacturer': healthfood['manufacturer'],
            'productType': healthfood['productType'],
            'mrpDecimal': healthfood['mrpDecimal'],
            'salePriceDecimal': healthfood['salePriceDecimal'],
            'discountPercent': healthfood['discountPercent'],
            'discountDecimal': healthfood['discountDecimal'],
            'categoryId': healthfood['categoryId'],
            'labelDetails': healthfood['labelDetails'],
            'ratingDetails': healthfood['ratingDetails'],
            'maxQuantity': healthfood['maxQuantity'],
            'productVolume': healthfood['productVolume'],
            'minQuantity': healthfood['minQuantity'],
            'isAvailable': healthfood['isAvailable'],
        }
        print(d)
        data.append(d)
```
